```python
import matplotlib.pyplot as plt

# Estimating the impact of the proposed changes
# Baseline data
total_orders = 300000
average_basket_size = 18.5
glovo_share = 0.30  # 30% share
uber_eats_share = 0.50  # 50% share
current_commission = 0.25  # 25%
proposed_commission = 0.20  # 20%

# Assuming a 10% increase in Glovo's orders due to improved promotions and reduced commission
new_glovo_orders = total_orders * (1+0.1)

# Assuming a 2% improvement in average basket size due to tiered discounts
new_average_basket_size = average_basket_size * 1.02

# Projected financials
current_revenue_glovo = total_orders * average_basket_size * current_commission
proposed_revenue_glovo = new_glovo_orders * new_average_basket_size *  0.23

# Glovo's new market share in deliveries
new_glovo_share = new_glovo_orders / total_orders

# Plotting the changes in share and financial impact
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Market Share Comparison
shares = [glovo_share, uber_eats_share]
labels = ['Glovo Current', 'Uber Eats Current']
colors = ['lightblue', 'lightgreen']
axs[0].pie(shares, labels=labels, autopct='%1.1f%%', colors=colors)
axs[0].set_title('Market Share Comparison')

# Financial Impact
revenues = [current_revenue_glovo, proposed_revenue_glovo]
labels = ['Current Revenue', 'Proposed Revenue']
axs[1].bar(labels, revenues, color=['red', 'green'])
axs[1].set_title('Financial Impact on Glovo')
axs[1].set_ylabel('Revenue (€)')
axs[1].set_ylim([min(revenues) - 50000, max(revenues) + 50000])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Estimating the impact of the proposed changes
# Baseline data
total_orders = 300000
average_basket_size = 18.5
glovo_share = 0.30  # 30% share
uber_eats_share = 0.50  # 50% share
current_commission = 0.25  # 25%
proposed_commission = 0.23  # 20%

# Assuming a 10% increase in Glovo's orders due to improved promotions and reduced commission
new_glovo_orders = total_orders * 1.10

# Assuming a 2% improvement in average basket size due to tiered discounts
new_average_basket_size = average_basket_size * 1.02

# Projected financials
current_revenue_burrito = total_orders * average_basket_size * (1-current_commission)
proposed_revenue_burrito = new_glovo_orders * new_average_basket_size *  (1-proposed_commission)

# Glovo's new market share in deliveries
new_glovo_share = new_glovo_orders / total_orders

# Plotting the changes in share and financial impact
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Market Share Comparison
shares = [glovo_share, uber_eats_share]
labels = ['Glovo Current', 'Uber Eats Current']
colors = ['lightblue', 'lightgreen']
axs[0].pie(shares, labels=labels, autopct='%1.1f%%', colors=colors)
axs[0].set_title('Market Share Comparison')

# Financial Impact
revenues = [current_revenue_glovo, proposed_revenue_glovo]
labels = ['Current Revenue', 'Proposed Revenue']
axs[1].bar(labels, revenues, color=['red', 'green'])
axs[1].set_title('Financial Impact on Burrito')
axs[1].set_ylabel('Revenue (€)')
axs[1].set_ylim([min(revenues) - 50000, max(revenues) + 50000])

plt.tight_layout()
plt.show()
```
